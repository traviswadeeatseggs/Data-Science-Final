## I did not remove any columns because they are needed to train a maching learning model. Additionally, the dataset as it is works perfectly to answer my first research question. However, I make a new dataset with the columns I need to answer question number two, since the current dataset wpuld be a pain to work with for that purpose.

```python
df2=pd.DataFrame()
```

```python
df2['mag']=df['mag'].round(1)
```

```python
df2['earthquakes_last_30_days']=df['earthquakes_last_30_days']
```

```python
df2['count']=df2['mag']
```

```python
df2['elapsed_time']=df['elapsed_time']
```

```python
df2=df2[df2['mag']>0]
```

```python
df3=df2.groupby('mag')['earthquakes_last_30_days'].mean().reset_index()
```

```python
df4=df2.groupby('mag')['count'].count().reset_index()
```

```python
df5=df2.groupby('mag')['elapsed_time'].mean().reset_index()
```

```python
df6=pd.merge(df3, df4, on='mag')
```

```python
df7=pd.merge(df6, df5, on='mag')
```

```python
#Downloading the new and improved dataset to use in R for the second research question
df7.to_csv('dfq1.csv')
```
